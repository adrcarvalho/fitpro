```python
!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata
api_key = "ENTER YOUR API"
genai.configure(api_key=api_key)

import textwrap
from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('•', ' *')
  return Markdown(textwrap.ident(text, '> ', predicate=lambda _: True))

generation_config = {
  "candidate_count": 1,
  "temperature": 0.5,
}

safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}

model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

print('Olá! Sou o FitPro, seu parceiro fitness virtual. Gostaria de criar um treino personalizado para você?')
print('\n')

chat = model.start_chat(history=[])

nome = input("Seu nome ")
idade = input("Sua idade ")
altura = input("Sua altura ")
peso = input("Seu peso ")

prompt = f"Meu nome é {nome}, tenho {idade} anos. Tenho {altura} de altura e peso {peso} kg. Crie um treino de academia para mim de 4 dias."

while prompt != "fim":
  response = chat.send_message(prompt)
  print("FitPro: ", response.text, "\n")
  prompt = input("Qual sua pergunta: ")
```
